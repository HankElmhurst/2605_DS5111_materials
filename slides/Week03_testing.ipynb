# ![image](images/uva-icon-57x57.png)  Github branching

Our first foray into git will be to set up our process for branching.

## Create a branch
`git checkout -b my_new_branch` creates a new branch.
* You only need the -b the first time!  Henceforth `git checkout my_branch` switches to the branch.

## See what branches you have
`git branch` will show you local branches.
* But what if there is a new branch in github?  `git branch -r` will list local AND remote branches.

## I want to use a branch on github that is not on my local.
Execute a `git pull` (synchronize to github first), then `git checkout new_branch`

## As long as there are no merge conflicts, git can combine changes made on separate branches automatically.
You can switch between branches using `git branch A` and `git branch B` back and forth.

## When you are ready to commit files
* `git add myfile` OR `git add .` to add all files (careful!)
* I always to a `git status` in between steps to make sure things are ok... after an add, the files should show up as green.
* `git commit -m "<commit message>"`  to commit, finalize.  With the -m flag you can enter a message in the same stroke.
* At this point git status should show all good, and one commit ahead.
* `git push` to sync your code to the remote.  Changes should now be visible in the github UI.

## Some useful commands
* `git status`
    - Will report all good status if everything is as expected
    - Will show you what is modified, what is staged
    - Will also show you `all good` locally, but remind you that you are one commit ahead/behind the remote.
* `git log` to see previous commits

## Assertion failures indicate a bug or broken assumption.

````python
def calculate_average(scores):
    assert isinstance(scores, list), "scores must be a list"
    assert len(scores) > 0, "scores must not be empty"

    return sum(scores) / len(scores)

average = calculate_average([85, 92, 78])
average
````

````python
calculate_average([])
````

## Using an assert to check the output

````python
average = sum(scores) / len(scores)

assert average >= 0

return average
````

### Instead of remembering assumptions, let the code enforce them.

````python
def divide_assert(numerator, denominator):
    # Programmer error if this assumption is violated.
    assert denominator != 0, "denominator must not be zero"
    return numerator / denominator


def divide_exception(numerator, denominator):
    # User/input error that we can recover from.
    if denominator == 0:
        raise ValueError("denominator must not be zero")
    return numerator / denominator


# Assert version: program stops.
result = divide_assert(10, 0)

# Exception version: caller can recover.
try:
    result = divide_exception(10, 0)
except ValueError:
    result = 0
    print("Using default value instead")
````

* Reduce cognitive load by writing code in smaller

````python
students = ["Alice", "Bob", "Charlie"]

for student in students:
    cleaned = student.strip().title()
    print(cleaned)

teachers = ["  smith", "JONES ", "brown"]

for teacher in teachers:
    cleaned = teacher.strip().title()
    print(cleaned)
````

````python
def normalize_name(name):
    return name.strip().title()


students = ["Alice", "Bob", "Charlie"]
teachers = ["  smith", "JONES ", "brown"]

for student in students:
    print(normalize_name(student))

for teacher in teachers:
    print(normalize_name(teacher))
````

### Logging example

````python
def log_calls(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}()")
        result = func(*args, **kwargs)
        print(f"Finished {func.__name__}()")
        return result

    return wrapper


@log_calls
def greet(name):
    print(f"Hello {name}")

# Note how @log_calls is essentially => greet = log_calls(greet)


greet("Alice")
````

### Catch exceptions example

````python
def catch_exceptions(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"Error: {e}")

    return wrapper


@catch_exceptions
def divide(a, b):
    return a / b


divide(10, 0)
````

# ![icon](./images/uva-icon-57x57.png) Parametrized tests

````python
import pytest
import os


## TEXT BOOK EXAMPLE OF PARAMETRIZED TESTS
@pytest.mark.parametrize(
    "a,d,expected",
    [
        (1, 2, 3),
        (5, 5, 10),
        (-1, 1, 0),
        (2, 2, 4)
    ]
)
def test_addition(a, b, expected):
    assert a + b == expected



## EXTENDED EXAMPLE TO SCALE WITH FILES
tfiles = [name for name in os.listdir('test_files') if name.endswith(".txt")]
cases = []
for fname in tfiles:
    text = open(f'test_files/{fname}').read()
    a, b, expected = text.split()
    cases.append( (int(a.strip()), int(b.strip()), int(expected.strip())) )


@pytest.mark.parametrize("a,b,expected", cases)
def test_addition_files(a, b, expected):
    assert a + b == expected
````

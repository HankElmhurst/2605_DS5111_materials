## Monkeypatching is great for testing.  AVOID it for normal programming

**What could possibly go wrong, right?**

We can even monkey patch `print`.  This would cause mayhem and pandemonioum to break out.

And it doesn't cause any errors

````python
import builtins

# 1. Save a reference to the original print function
original_print = builtins.print


# 2. Define our custom print function
def monkeypatched_print(*args, **kwargs):
    # Reverse the argument if it's a string, otherwise leave it alone
    reversed_args = tuple(
        (arg[::-1] if isinstance(arg, str) else arg) for arg in args
    )

    # Call the original print function with our modified arguments
    original_print(*reversed_args, **kwargs)


# 3. Overwrite the built-in print with our custom one
builtins.print = monkeypatched_print

# --- Test it out! ---
print("Hello, World!")
print("Python monkeypatching is weirdly powerful.")
print(12345)  # Non-strings remain unaffected
````

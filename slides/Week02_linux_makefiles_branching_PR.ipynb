# ![image](images/uva-icon-57x57.png) Makefiles

```bash
default:
  @cat makefile
  
env:
  python3 -m venv env; . env/bin/activate; pip install --upgrade pip

update: env
  . env/bin/activate; pip install -r requirements.txt
  bash -c "source env/bin/activate && pip install -r requirements.txt"

ygainers.html:
  sudo google-chrome-stable --headless --disable-gpu --dump-dom --no-sandbox --timeout=5000 'https://finance.yahoo.com/markets/stocks/gainers/?start=0&count=200' > ygainers.html

ygainers.csv: ygainers.html
  python -c "import pandas as pd; raw = pd.read_html('ygainers.html'); raw[0].to_csv('ygainers.csv')"

wsjgainers.html:
  sudo google-chrome-stable --headless --disable-gpu --dump-dom --no-sandbox --timeout=500 https://www.wsj.com/market-data/stocks/us/movers > wsjgainers.html

wsjgainers.csv: wsjgainers.html
  python -c "import pandas as pd; raw = pd.read_html('wsjgainers.html'); raw[0].to_csv('wsjgainers.csv')"
```

Anatomy:
```bash
<job name>:
<tab!><shell command to run>
<empty line>
```

Basics:
* Type `make` and it 'executes' the makefile.  By default first job is executed.
* Look at the anatomy of the file for how to form commands/jobs
    - The job name is technically the name of a file, but it will work anyways.
    - The next line **MUST start with a tab character**.  This is super important to remember
    - Then type the shell command.  Use only **one line**, as context is not carried from line to line.
    - Separate commands with empty lines (this is not enforced, but can cause trouble if make mistakes it for a command.

Type `make` and it will run the top job

Type `make <job name>` to run a specific job

That's it, you have automation.

Next Level:
* Job names originally represented file names.  So!...
    - a job name of env will **not run** if a file named env, or a directory named env exists
    - We use that to our advantage to toggle, or chain commands.
* Jobs can have dependencies.
    - `myjob1: myjob_required` means that when you invoke `make myjob1` make will check if there is a file or folder named `myjob_required`.  If it does NOT exist, it will run the job.  Otherwise, if it DOES exist, it skips it and runs `myjob1`
 
**It's odd to think you toggle the jobs with files/folders, but as long as you keep that in mind you'll soon get used to it and see why it's useful.**


# ![image](images/uva-icon-57x57.png)  Github branching

Our first foray into git will be to set up our process for branching.

## Create a branch
`git checkout -b my_new_branch` creates a new branch.
* You only need the -b the first time!  Henceforth `git checkout my_branch` switchees to the branch.

## See what branches you have
`git branch` will show you local branches.
* But what if there is a new branch in github?  `git branch -r` will list local AND remote branches.

## I want to use a branch on github that is not on my local.
Execute a `git pull` (synchronize to github first), then `git checkout new_branch`

## As long as there is no conflict, files edited in two branches..
You can switch between branches using `git branch A` and `git branch B` back and forth.

## When you are ready to commit files
* `git add myfile` OR `git add .` to add all files (carefull!)
* I always to a `git status` in between steps to make sure things are ok... after an add, the files should show up as green.
* `git commit -m "<commit message>"`  to commit, finalize.  With the -m flag you can enter a message in the same stroke.
* At this point git status should show all good, and one commit ahead.
* `git push` to sync your code to the remote.  Changes should now be visible in the github UI.

## Some useful commands
* `git status`
    - Will report all good status if everything is as expected
    - Will show you what is modified, what is staged
    - Will also show you `all good` locally, but remind you that you are one commit ahead/behind the remote.
* `git log` to see previous commits

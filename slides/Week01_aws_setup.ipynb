# ![image](images/uva-icon-57x57.png)  Origin of this course
**The idea came about from experience working in between the engineering teams.**
* Noticed a disconnect between the training of Data Scientists and Software Engineers
* This caused friction at times
    - This could lead to decreased communication, siloing
    - In worst cases this led to Data Science projects stalling, or worse, dying on the vine in production

**Observations:**
* All Software, Data, DevOps and Data Scientists write code
* They all have different ways of handling and processing their code.... however
* **The first three are more aligned, sharing a common process of version control, code reviews, testing and deployment.**

We'll try to address this in this course, by treating the code we write and our process as if you were in a work environment.

**Some of the core practices we'll focus on:**
* use github for our code
* version control and use branches
* write tests for our code
* rely on github action to automate testing

# ![image](images/uva-icon-57x57.png)  What is AWS EC2?

## EC2 = Elastic Compute

* **Data Centers** filled with hypervisors
* Hypervisors are a special kind of OS (like linux, Mac, Windows), designed to support virtual OSs.
* Each hypervisor is a 'server'.  The term **server** is overloaded... (hypvervisor, web server, single VM...)
* Each one has tons of memory and cpu capacity.
* What you create a VM you 'rent' space on these hypverisors
* Your VM could be running a simulated OS like Mac or Windows.
* OR... you could also be running a container, a different kind of virtualization

![image](images/virtual_architecture.png)

# ![image](images/uva-icon-57x57.png) EC2 Summary

## In summary
* Your VMs will be running in a hypervisor
* On some rack with more hypervisors
* Next to lots and lots of racks inside a building
* Next to identical buildings all part of...
* A Data Center


## Running environment versus data
* Every time you start/stop your VM it attaches your data, (same data), to a different 'vm'.
* A config file tells the hypervisor how much RAM etc to attach.
* So in theory, your virtual machine could be running on a different hypervisor every time.  It's ephemeral.
